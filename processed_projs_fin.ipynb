```python
##Create dictionaries to map the dimension ids to fact table
dict_year = {year_name[i]:year_id[i] for i in range(len(year_name))}
dict_donor = {donor_name[i]:donor_id[i] for i in range(len(donor_name))}
dict_recipient = {recipient_name[i]:recipient_id[i] for i in range(len(recipient_name))}
dict_recip_region = {recipient_region[i]:recipient_region_id[i] for i in range(len(recipient_region))}
```

```python
###Map dimension ids
fct_projects.rename(columns={'year':'year_name'},inplace=True)
fct_projects.head()
fct_projects['year_name']= fct_projects['year_name'].map(dict_year)
fct_projects['donor_name']= fct_projects['donor_name'].map(dict_donor)
fct_projects['recipient_name']=fct_projects['recipient_name'].map(dict_recipient)
fct_projects['recipient_region'] = fct_projects['recipient_region'].map(dict_recip_region)
fct_projects.head()
```

```python
fct_projects.rename(columns={'year_name':'year_id','donor_name':'donor_id',
                              'recipient_name':'recipient_id',	'recipient_region':'recipient_region_id'}
                     ,inplace=True)
fct_projects.head()
```

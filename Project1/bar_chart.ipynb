```python
import matplotlib.pyplot as plt
```

```python
employee_data = {
    "Alterations": 73,
    "Assistant Manager": 81,
    "Barista": 94,
    "Bartender": 79,
    "Beauty Concierge": 89,
    "Cafe Cashier": 79,
    "Counter Manager": 86,
    "Food Server": 64,
    "Manager": 68,
    "Retail Cashier & Sales Floor Support": 67,
    "Retail Stock": 76,
    "Sales Associate": 76,
    "Stylist": 69
}
```

```python
positions = list(employee_data.keys())
```

```python
counts = list(employee_data.values())
```

```python
plt.figure(figsize=(10, 6))
plt.bar(range(len(employee_data)), counts, tick_label=positions, color='orange')
plt.xticks(rotation=45, ha='right')
plt.xlabel("Employee Title")
plt.ylabel("Employee Count")
plt.title("Employee Count by Title")
plt.tight_layout()
plt.show()
```
